```python
import pandas as pd
import numpy as np

# Create the DataFrame
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=labels)

# Select specific columns and rows
selected_df = df.loc[['b', 'd', 'f', 'g'], ['name', 'score']]

# Print the selected DataFrame
print("Select specific columns and rows:")
print(selected_df)
```

2. Use Crime dataset from LMS

I) find the aggregations like all moments of business decisions for all columns,value counts.

II) do the plottings like plottings like histogram, boxplot, scatterplot, barplot, piechart,dot chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
crime_data = pd.read_csv('crime_data.csv')  # Replace 'crime_data.csv' with your dataset filename or path

# Step 2: Display the data and basic information
print("Data sample:")
print(crime_data.head())

print("\nData summary:")
print(crime_data.describe())

print("\nData information:")
print(crime_data.info())

# Step 3: Perform aggregations using all moments of business decisions
aggregations = crime_data.agg(['mean', 'median', 'min', 'max', 'std', 'var'])
print("\nAggregations using all moments of business decisions:")
print(aggregations)

# Step 4: Value counts for all columns
value_counts = {}
for column in crime_data.columns:
    value_counts[column] = crime_data[column].value_counts()
    print("\nValue counts for column", column, ":")
    print(value_counts[column])

# Step 5: Plottings

# Histogram
crime_data['Rape'].plot.hist(bins=20)
plt.xlabel('Unnamed: 0')
plt.ylabel('Frequency')
plt.title('Unnamed: 0 Histogram')
plt.show()

# Boxplot
sns.boxplot(x='Rape', y='Murder', data=crime_data)
plt.xlabel('Year')
plt.ylabel('Violent crime total')
plt.title('Violent Crime Boxplot')
plt.show()

# Scatterplot
sns.scatterplot(x='Rape', y='Murder', data=crime_data)
plt.xlabel('Rape')
plt.ylabel('Murder')
plt.title('Rape vs. Murder')
plt.show()

# Barplot
crime_data['Unnamed: 0'].value_counts().plot.bar()
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.title('Crime Data by Region')
plt.show()

# Pie chart
crime_data['UrbanPop'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('UrbanPop')
plt.axis('equal')
plt.show()

```

4. Use Bank Dataset from LMS

A)change all the categorical columns into numerical by creating Dummies and using label encoder.

B) rename all the column names DF

C) Rename only one specific column in DF

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the Bank dataset
bank_data = pd.read_csv('bank-full.csv')  #Replace 'bank_data.csv' with your dataset filename or path

# Step 2A:Convert categorical columns to numerical using dummies (One-Hot Encoding)
bank_dummies = pd.get_dummies(bank_data)
print("Dataset after creating dummies:")
print(bank_dummies)

# Step 2A: Convert categorical columns to numerical using LabelEncoder
bank_encoded = bank_data.copy()
label_encoder = LabelEncoder()
for column in bank_encoded.columns:
    if bank_encoded[column].dtype == 'object':
        bank_encoded[column] = label_encoder.fit_transform(bank_encoded[column])
print("\nDataset after Label Encoding:")
print(bank_encoded)

# Step 3: Rename all column names in the DataFrame
new_columns = ['DF' + str(i) for i in range(len(bank_data.columns))]
bank_renamed_all = bank_data.copy()
bank_renamed_all.columns = new_columns
print("\nDataset with renamed column names (all columns):")
print(bank_renamed_all)

# Step 4: Rename a specific column in the DataFrame
bank_renamed_specific = bank_data.copy()
bank_renamed_specific = bank_renamed_specific.rename(columns={'specific_column_name': 'new_column_name'})
print("\nDataset with renamed specific column:")
print(bank_renamed_specific)
```
